Part 4: Real-world Application and Impact
1. Business ROI Analysis (Text)
By implementing the ANN for Customer Churn (Part 1) and CNN for Visual Inspection (Part 2), we can observe a significant improvement in business efficiency.

Efficiency Gain: Automating the identification of "at-risk" customers and "defective" products reduces manual review time.

Cost Reduction: Based on the business_kpi_sample.csv data, there is a clear trend showing that as AI models are integrated, Manual Processing Hours and Error Rates drop significantly.

Customer Satisfaction: Predicting churn allowed the bank to offer proactive discounts, leading to a higher customer satisfaction score.

2. Responsible AI & Risk Assessment (Text)
Using the ai_usecase_reference_catalog.csv as a guide, we identified the following risks for our models:

For Part 1 (Churn): The risk is Targeting customers unfairly or privacy concerns regarding financial data. We must ensure the model doesn't discriminate based on geography or gender.

For Part 2 (Visual Inspection): The risk involves Missed defects due to lighting or camera quality issues, which could lead to shipping faulty products.

```python
import pandas as pd
import matplotlib.pyplot as plt
from google.colab import files

# 1. Load the provided reference documents
kpi_df = pd.read_csv('business_kpi_sample.csv')
catalog_df = pd.read_csv('ai_usecase_reference_catalog.csv')

# 2. GENERATE BUSINESS ROI CHART (Manual Hours vs Month)
plt.figure(figsize=(12, 6))
plt.plot(kpi_df['month'], kpi_df['manual_processing_hours'], marker='o', color='#e67e22', label='Manual Processing Hours', linewidth=3)
plt.plot(kpi_df['month'], kpi_df['error_rate_percent'] * 50, marker='s', color='#3498db', label='Error Rate (Scaled x50)', linestyle='--')

plt.title('Business ROI: Impact of AI Implementation Over Time', fontsize=14, fontweight='bold')
plt.xlabel('Month (2025)', fontsize=12)
plt.ylabel('Hours / Error Magnitude', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.3)

plt.savefig('business_roi_impact.png', bbox_inches='tight')
plt.show()

# 3. GENERATE RESPONSIBLE AI RISK ASSESSMENT (Extracting from Catalog)
# Filter for Customer Churn (Part 1) and Manufacturing (Part 2)
churn_info = catalog_df[catalog_df['business_problem'] == 'Customer churn prediction']
vision_info = catalog_df[catalog_df['business_problem'] == 'Visual defect inspection']

responsible_ai_data = {
    'Project Part': ['Part 1: ANN (Churn)', 'Part 2: CNN (CV)'],
    'AI Task': [churn_info['ai_task_type'].values[0], vision_info['ai_task_type'].values[0]],
    'Responsible AI Risk': [churn_info['responsible_ai_risk'].values[0], vision_info['responsible_ai_risk'].values[0]]
}

risk_df = pd.DataFrame(responsible_ai_data)
risk_df.to_csv('responsible_ai_risks.csv', index=False)

print("\n--- Responsible AI Risk Assessment ---")
print(risk_df)

# 4. DOWNLOAD THE FILES
print("\nDownloading Part 4 Files...")
files.download('business_roi_impact.png')
files.download('responsible_ai_risks.csv')
```
